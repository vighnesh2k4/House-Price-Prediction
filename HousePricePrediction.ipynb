```python
#Importing libraries required for this particular task

import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, accuracy_score
from sklearn.preprocessing import MinMaxScaler
```

```python
#Importing the Housing dataset

housedata = pd.read_csv("Housing.csv")
```

```python
#Projecting the first 10 rows of the dataset are as follows:

housedata.head(10)
```

```python
#The number of rows and columns in the dataset are given by the shape attribute

housedata.shape
```

```python
#The summary statistics of the dataset are as follows:

housedata.describe()
```

```python
#Applying Min-Max scaling to the columns of the dataset

columns = ['price', 'area', 'bedrooms', 'bathrooms', 'stories']
scale = MinMaxScaler()
housedata[columns] = scale.fit_transform(housedata[columns])
```

```python
#Converting Categorical Values into Numerical Format using Label Encoding

categories = ["mainroad","guestroom","basement","hotwaterheating","airconditioning","prefarea","furnishingstatus"]


label = LabelEncoder()
housedata[categories] = housedata[categories].apply(label.fit_transform)
```

```python
#The first 10 rows of the dataset after applying Label Encoding and Min-Max Scaling are as follows:

housedata.head(10)
```

```python
#Identifying the missing values in the dataset

housedata.isnull().sum()
```

```python
#Looking out for duplicates
#Making the dataset free of duplicate entries

dup = housedata[housedata.duplicated(keep=False)]
num_dup = dup.shape[0]

if num_dup == 0:
    print("This dataset doesn't contain any duplicates")
else:
    housedata.drop_duplicates(inplace=True)
    print("Duplicate entries are Dropped")
```

```python
#Making use of a heatmap to visualize the correlation between features

plt.figure(figsize=(15,15))
sb.heatmap(housedata.corr(), cbar=True, square=True, fmt = '.1f', annot = True, annot_kws={'size':8})
```

```python
#Visualizing relationships between different features by making use of a pairplot

sb.pairplot(housedata)
```

```python
#Visualizing the distribution of each feature by using Histograms

housedata.hist(figsize=(15,13),bins=50)
plt.show()
```

```python
#The Independent Variables (x) and Dependent Variable (y) of the dataset

x = housedata.drop('price',axis=1)
y = housedata['price']
```

```python
x.head()
```

```python
y.head()
```

```python
#Constructing training and testing sets by splitting the dataset

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#Making use of a Linear Regression model

linear_model = LinearRegression()
linear_model.fit(x_train, y_train)

#Making predictions
linear_y_pred = linear_model.predict(x_test)
```

```python
#Finding out evaluation metrics for the Linear Regression model

linear_r2 = r2_score(y_test, linear_y_pred)
linear_mae = mean_absolute_error(y_test, linear_y_pred)
linear_mse = mean_squared_error(y_test, linear_y_pred)
linear_rmse= mean_squared_error(y_test, linear_y_pred, squared=False)

print(f'Coefficient of Determination or R-squared value : {linear_r2*100}')
print(f'Mean Absolute Error value : {linear_mae}')
print(f'Mean Squared Error value : {linear_mse}')
print(f'Root Mean Squared Error value : {linear_rmse}')
```

```python
#Making use of Scatter Plot
#Plotting the Actual Values against the Predicted Values

plt.figure(figsize=(8, 6))
plt.scatter(y_test, linear_y_pred, c='purple', alpha=0.6)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual Values against the Predicted Values ")
plt.text(0.1, 0.9, f'R-squared : {linear_r2*100:.2f}', transform=plt.gca().transAxes)
plt.grid(True)
plt.show()
```
